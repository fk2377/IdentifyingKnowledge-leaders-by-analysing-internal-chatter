```python
def business_keywords(msg1):
    #print(msg1)
    list1=[]
    tokens=nltk.word_tokenize(msg1)
    exclude = set(string.punctuation) # remove punctation marks from the message
    tokens1=[ch for ch in tokens if ch not in exclude]
    for i in range(0,len(tokens1)):
        #print(tokens[i].lower())                    
        if tokens1[i].lower() in techwords :
            list1.append(tokens1[i].lower())
    if len(list1)>0:
        return " ".join(list1)
    else: return np.NaN
```

```python
def filter_startword(startword):
    
    try :
        startword=startword.lower()
        df5=df4[df4['startword']==startword]



        for row in df5.iterrows():
            index, data = row
            temp=(data.tolist())

        temp.pop(0)



        return [x for x in temp if x != 'buriub']  
    
    except:
        return ('NA')
  #
```

```python
from matplotlib import pylab
import networkx as nx

def save_graph(graph,file_name):
    #initialze Figure
    plt.figure(num=None, figsize=(20, 20), dpi=80)
    plt.axis('off')
    fig = plt.figure(1)
    pos = nx.spring_layout(graph)
    nx.draw_networkx_nodes(graph,pos)
    nx.draw_networkx_edges(graph,pos)
    nx.draw_networkx_labels(graph,pos)

    cut =1.5
    xmax = cut * max(xx for xx, yy in pos.values())
    ymax = cut * max(yy for xx, yy in pos.values())
    
    xmin = cut * min(xx for xx, yy in pos.values())
    ymin = cut * min(yy for xx, yy in pos.values())
    
    plt.xlim(xmin, xmax)
    plt.ylim(ymin, ymax)
    #plt.show()
    #plt.savefig('X:/' + file_name + '.png', bbox_inches='tight')
    plt.savefig(file_name,bbox_inches="tight")
    pylab.close()
    del fig

#Assuming that the graph g has nodes and edges entered
save_graph(G,"all_questionsNW.pdf")
```
