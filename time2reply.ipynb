```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('tech_Score_added.csv')
```

```python
df
```

```python
df_time=df[['user_id','dest','time_reply']]
```

```python
sf=df['dest'].value_counts()
```

```python
df1=pd.DataFrame({'user_id':sf.index, 'count':sf.values})
```

```python
df1=df1[['user_id','count']]
```

```python
def avg_reply_time(name_of_replier,non_outlier_part_percentage):
    booleans = []
    for start in df_time.dest:
        
        #print(start)
        if start == name_of_replier:
                booleans.append(True)
        else:
            booleans.append(False)
        
    Specifichour = pd.Series(booleans)

    df6=df_time[Specifichour]
    df6=df6.sort_values(by='time_reply', ascending=True)
    avg_rep_time=df6[:round(len(df6)*non_outlier_part_percentage)]['time_reply'].mean()
    
    return avg_rep_time
```

```python
'''avg_reply_time('U02EBRG16',100/100)'''
```

```python
user_id_of_replier=df_time['dest'].tolist()
```

```python
#user_id_of_replier=['U02EBRG16']
```

```python
list_60_avg_time=[]

list_70_avg_time=[]


list_90_avg_time=[]

for i in user_id_of_replier:
    a=avg_reply_time(i,60/100)
    list_60_avg_time.append(a)
    
for i in user_id_of_replier:
    a=avg_reply_time(i,70/100)
    list_70_avg_time.append(a)
    
for i in user_id_of_replier:
    a=avg_reply_time(i,90/100)
    list_90_avg_time.append(a)
```

```python
avg_time_df = pd.DataFrame(np.column_stack([user_id_of_replier,list_60_avg_time, list_70_avg_time, list_90_avg_time]), 
                               columns=['user_id', 'list_60_avg_time','list_70_avg_time', 'list_90_avg_time'])
```

```python
avg_time_df 
```

```python
merge_avg_time_df =pd.merge(df1, avg_time_df, on='user_id', how='inner')
#merge_avg_time_df.to_csv('timetoansweravg_60_70_90.csv')
#del merge_avg_time_df['level_0']
```
